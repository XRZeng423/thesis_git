```python
import pandas as pd
import numpy as np
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
import numpy as np

finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone',num_labels=3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')

```

```python
text = pd.read_csv('/home/xuranzeng/thesis/text_path.csv').iloc[: , 1:]
```

```python
text.head()
```

```python
for i in range(len(text)):
#for i in range(2):

    text_path = text.loc[i,'Complete_Path']
    
    # open text
    with open(text_path) as f:
        contents = f.read()
        sentences = contents.split('\n')[:-1]  
        
    # label
    inputs = tokenizer(sentences, return_tensors="pt", padding=True)
    outputs = finbert(**inputs)[0]

    labels = {0:'neutral', 1:'positive',2:'negative'}

    sentence = []
    label = []
    for idx, sent in enumerate(sentences):

        sentence.append(sent)
        label.append(labels[np.argmax(outputs.detach().numpy()[idx])])
        #print(sent, '----', labels[np.argmax(outputs.detach().numpy()[idx])])

    label_text = pd.DataFrame({'Sentences':sentence,'Labels':label})

    # save
    base_path = text_path[:-8]
    name = 'label_text.csv'

    save_path = base_path + name
    label_text.to_csv(save_path)
    
    print('processing:',i,save_path)
```
